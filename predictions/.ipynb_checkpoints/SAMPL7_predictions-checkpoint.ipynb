```python
import pandas as pd
from rdkit.Chem import PandasTools
from matplotlib import pyplot as plt

#db = pd.read_pickle("pKaDatabase/Database_final.pkl")
training = pd.read_pickle("pKaDatabase/Database_final.pkl")
dataset = pd.read_pickle("pKaDatabase/Sulfonamides.pkl")
db = pd.concat([training, dataset])



file = "k_fold_cross_val_results_norm_ochem_final_937_with_sulfonamides.pkl"

PandasTools.AddMoleculeColumnToFrame(db, smilesCol='protonated microstate smiles', molCol='protonated molecule')
PandasTools.AddMoleculeColumnToFrame(db, smilesCol='deprotonated microstate smiles', molCol='deprotonated molecule')

import io
buffer = io.StringIO()
db.info(buf=buffer)
s = buffer.getvalue()
info = s.split("\n")
values = []
for i in range(10,33):
    values.append(int(info[i].split("non-null")[0].split("  ")[-1]))
print(f'Molecules with features: {min(values)}/{len(db.index)} = {min(values)/len(db.index)}')
print(db.info())
#db


feats = ['AM1BCC partial charge (prot. atom)', 'AM1BCC partial charge (deprot. atom)',
         'AM1BCC partial charge (prot. atoms 1 bond away)',
         'AM1BCC partial charge (deprot. atoms 1 bond away)',
         'AM1BCC partial charge (prot. atoms 2 bond away)',
         'AM1BCC partial charge (deprot. atoms 2 bond away)',
         '∆G_solv (kJ/mol) (prot-deprot)', 'SASA (Shrake)', 'Bond Order',
         'Change in Enthalpy (kJ/mol) (prot-deprot)', 'pKa']
training = db[feats]

feature_list=['PC (AH)', 'PC (A)', 'PC (AH 1bond)', 'PC (A 1bond)',
    'PC (AH 2bond)', 'PC (A 2bond)', r'$∆G_{solv}$(AH-A) (kJ/mol)',
    'SASA', 'Bond Order', '∆H(AH-A) (kJ/mol)', 'pKa']

training.columns = feature_list
training_ = training
results = pd.read_pickle(file)
fig = SAMPL7.GPR.corner_plot(training_)

#fig.savefig("corner_plot.pdf", dpi=400)
fig.set_figheight(16);fig.set_figwidth(16)
#fig.savefig("corner_plot_training.pdf", dpi=400)
fig.show()

fig = plt.figure()
ax = plt.subplot(2,2,1)
ax1 = plt.subplot(2,2,3)
ax2 = plt.subplot(1,2,2)
db.sort_values('Weight', ascending=True)
ax = db["Weight"].plot.hist(rot=0, figsize=(10, 4), bins=25, edgecolor='black', linewidth=1.2, ax=ax)
ax.set_xlabel("MWt", size=16)
ax.set_ylabel("")
ax.axvline(x=db['Weight'].mean(), linewidth=4, color='r')

db = training#pd.read_pickle("training_set_subset.pkl")
db = db.dropna()

ax1 = db["pKa"].plot.hist(rot=0, figsize=(10, 4), bins=25, edgecolor='black', linewidth=1.2, ax=ax1)#, subplots=True, layout=(2,2))
ax1.set_xlabel(r"$pK_{a}$", size=16)
ax1.set_ylabel("")
#ax1.axvline(x=4, linewidth=4, color='r')
#ax1.axvline(x=9, linewidth=4, color='r')
###########################################
#file = "k_fold_cross_val_results_norm_ochem_4_fold.pkl"
results = pd.read_pickle(file)
posterior = np.concatenate(list(results["posterior"]))
y_test = np.concatenate(list(results["y_test"]))
lower_lim, upper_lim = -3, 15
limit = np.array([lower_lim, upper_lim])
ax2.scatter(y_test, posterior )
ax2.plot(limit, limit, color="k")
ax2.fill_between(limit, limit+1.0, limit-1.0, facecolor='brown', alpha=0.35)
ax2.fill_between(limit, limit+2.0, limit-2.0, facecolor='wheat', alpha=0.50)
ax2.set_xlabel("Observed", size=16)
ax2.set_ylabel("Predicted", size=16)
ax2.set_xlim(limit)
ax2.set_ylim(limit)
posx,posy = np.min(y_test)+5.1, np.max(posterior) - 4
string = r'''$N$ = %i
$R^{2}$ = %0.3f
MAE = %0.3f
MSE = %0.3f
RMSE = %0.3f'''%(len(posterior), results.mean()["R^2"], results.mean()["MAE"], results.mean()["MSE"], results.mean()["RMSE"])
ax2.annotate(string, xy=(posx,posy), xytext=(posx,posy))
#ax2.axvline(x=4, linewidth=4, color='r')
#ax2.axvline(x=9, linewidth=4, color='r')
###########################################
fig = ax1.get_figure()
fig.tight_layout()
#fig.savefig("OChemDatabase_stats.pdf")
#fig.savefig("OChemDatabase_stats.pdf")
```

```python
import SAMPL7
import pandas as pd
from IPython.display import display, HTML
from IPython.display import Image
pd.options.display.max_columns = 50
pd.options.display.max_rows = 20

#network_imgs = SAMPL7.toolbox.get_files("Structures/SAMPL7/input_files/SM*/Network_of_states.png")                                        
input_files = SAMPL7.toolbox.get_files("Structures/SAMPL7/input_files/SM*/SM*.csv")
feature_files = SAMPL7.toolbox.get_files("SM*/total_feature_input.pkl")
results_files = SAMPL7.toolbox.get_files("SM*/*_results.pkl")
transitions_files = SAMPL7.toolbox.get_files("Structures/SAMPL7/input_files/SM*/SM*_transitions.npy")

for i in range(len(input_files)):
    SM = results_files[i].split("/")[0]
    SAMPL7.Processing.graph_from_df(input_files[i], transitions_files[i], 4, results_files[i],
                                    figname=SM+f"/{SM}_network.png", edge_label_font_size=14,
                                    molSize=0.20, subImgSize=(155,155), figsize=(6, 10))

import numpy as np    
macro_pKas = []
for i,input_file in enumerate(input_files):
    SM = results_files[i].split("/")[-2]
    #print(f'Small Molecule: {SM}')
    in_file = pd.read_csv(input_file, index_col=False)
    #print(in_file)
    max_charge = np.max(list(in_file["charge state"]))
    min_charge = np.min(list(in_file["charge state"]))
    charge = max_charge
    pKas = {}
    pKas["Molecule"] = SM
    while charge >= min_charge+1:
        #print(input_file, results_files[i], feature_files[i])
        pKas[charge] = SAMPL7.Processing.get_macroscopic_pKas(int(charge), input_file, results_files[i], feature_files[i])
        charge -= 1
    macro_pKas.append(pKas)
macro_pKas = pd.DataFrame(macro_pKas)
macro_pKas = macro_pKas[[0,1,2]]
macro_pKas.to_csv("macropKa_predictions.csv", index=False)
display(HTML(macro_pKas.to_html()))


Gs = []
for i,input_file in enumerate(input_files):
    SM = results_files[i].split("/")[-2]
    in_file = pd.read_csv(input_file, index_col=False)
    free_energies = SAMPL7.Processing.get_Gs("micro000", input_file, results_files[i], feature_files[i])
    free_energies = SAMPL7.Processing.fix_Gs(free_energies, input_file, results_files[i], feature_files[i])
    Gs.append(free_energies)
Gs = pd.concat(Gs)
col_order = ['Ref', 'Microstate ID', 'Charge', '∆G', '∆G std']
Gs = Gs[col_order]
Gs.to_csv("free_energy_predictions.csv", index=False)
display(HTML(Gs.to_html())) 
    
    
network_imgs = SAMPL7.toolbox.get_files("SM*/SM*_network.png")                                         
for i in range(len(input_files)):
    print(f'Small Molecule: {network_imgs[i].split("/Network_of_states.png")[0].split("/")[-1]}')
    display(Image(network_imgs[i]))
    df = pd.read_csv(input_files[i])
    display(HTML(df.to_html()))
    df2 = pd.read_pickle(feature_files[i])
    df1 = pd.read_pickle(results_files[i])
    df = pd.concat([df1,df2], axis=1)
    display(HTML(df.to_html()))
    print("\n\n")
```
