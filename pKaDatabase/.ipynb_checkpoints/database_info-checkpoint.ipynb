```python
import pandas as pd
import numpy as np
from rdkit import Chem
import sys
sys.path.append("../")
import SAMPL7
from IPython.display import display, HTML
```

```python

def get_training_from_subset(subset, df):
    """Find all the matches for each molecule in the subset of the dataset
    """
    
    N = len(list(subset["pKa"]))
    matches = []
    for i in range(N):
        match = np.where((list(subset["pKa"])[i]==df["pKa"]) & \
                         (list(subset["Bond Order"])[i]==df["Bond Order"]) & \
                         (list(subset["AM1BCC partial charge (prot. atom)"])[i]==df["AM1BCC partial charge (prot. atom)"]) & \
                         (list(subset["AM1BCC partial charge (deprot. atom)"])[i]==df["AM1BCC partial charge (deprot. atom)"]))[0]
        
        
        #print(match)
        if match == []: print("No match. Exit.");exit();
        if len(match) > 1: print("More than 1 match. Exit.");exit();
        else: matches.append(int(match[0]))
    #print(matches)
    ans = df.iloc[list(matches)]
    return ans
    
def get_smiles_from_input_files(input_files):
    # get all of the molecules of interest
    
    all_smiles = []
    for file in input_files:
        df = pd.read_csv(file)
        SM = file.split("/")[-1].split(".csv")[0]
        smiles_list = list(df["smiles"])
        for smiles in smiles_list:
            all_smiles.append([SM,smiles])
    return all_smiles

            
input_files = SAMPL7.toolbox.get_files("/Users/rr/github/SAMPL7/Structures/SAMPL7/input_files/*/SM*.csv")
microstate_smiles = get_smiles_from_input_files(input_files)
traning_set_files = SAMPL7.toolbox.get_files("/Users/rr/github/SAMPL7/predictions/model_validation/*/*/training_set_*.pkl")
#print(traning_set_files)

container = []
for file in traning_set_files:
    proticity = file.split("/")[-3]
    nMolecules = file.split("/")[-1].split("training_set_k_fold_cross_val_")[-1].split(".")[0].split("_")[0]
    training_set = pd.read_pickle(file)
    training = get_training_from_subset(subset=training_set, df=total_dataset)
    ts = SAMPL7.RD.get_fingerprints(training, "protonated microstate smiles")
    averages = []
    for SM,smiles in microstate_smiles:
        sim = SAMPL7.RD.get_similarity(ts, moi_smiles=smiles)
        avg = sim["Fingerprint Similarity"].mean()
        container.append({"nMolecules": nMolecules, "Group": proticity,
                         "Avg. Fingerprint Similarity": avg, "Small Molecule": SM, "Smiles": smiles})
fp_analysis = pd.DataFrame(container)
fp_analysis
    
```
